```python
import mysql.connector##Importing required libraries
from mysql.connector import Error
import pandas as pd
```

```python
def create_server_connection(host_name, user_name, user_password):#Creating a method to to connect to the server
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
pw = "mysqlpass"
connection = create_server_connection("localhost", "root", pw)##Creating server connection
```

```python
def create_database(connection, query):##Creating a method to thats handles the database creation query
    cursor = connection.cursor()##Assigning cursor
    try:
        cursor.execute(query)##Asking cursor to execute query
        print("Database created successfully")
    except Error as err: ##If error print error
        print(f"Error: '{err}'")
def create_db_connection(host_name, user_name, user_password, db_name):## Creating a method to connect to the database
    connection = None
    try:
        connection = mysql.connector.connect( ##Connection to MySQL
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection 
def execute_query(connection, query):## This will handle and send queries to the cursor for execution in mysql
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
# Creating Tables in normalized form as made in report
create_property_table = """
CREATE TABLE property (
  prop_id INT PRIMARY KEY, 
  prop_address VARCHAR(40) NOT NULL,
  prop_price INT,
  prop_availability VARCHAR(40) NOT NULL
  );
 """ 

create_customer_table = """
CREATE TABLE customer (
  cust_id INT PRIMARY KEY,
  cust_name VARCHAR(40) NOT NULL
  );
 """
create_agency_table = """
CREATE TABLE agency(
ag_id INT PRIMARY KEY,
ag_name VARCHAR(40) NOT NULL
);
"""
create_enlists_table = """
CREATE TABLE enlists(
cust_id INT,
ag_id INT,
FOREIGN KEY(cust_id) REFERENCES customer(cust_id),
FOREIGN KEY(ag_id) REFERENCES agency(ag_id)
);
"""
create_agent_table = """
CREATE TABLE agent(
a_id INT PRIMARY KEY,
a_name VARCHAR(40) NOT NULL
);
"""
create_assigns_table = """
CREATE TABLE assigns(
ag_id INT,
a_id INT,
FOREIGN KEY(ag_id) REFERENCES agency(ag_id),
FOREIGN KEY(a_id) REFERENCES agent(a_id)
);
"""
create_seller_table = """
CREATE TABLE seller(
seller_no INT PRIMARY KEY,
seller_name VARCHAR(40) NOT NULL
);
"""
create_searches_table = """
CREATE TABLE searches(
agent_id INT,
s_no INT,
FOREIGN KEY(agent_id) REFERENCES agent(a_id),
FOREIGN KEY(s_no) REFERENCES seller(seller_no)
);
"""
create_sells_table = """
CREATE TABLE sells(
seller_id INT,
seller_name VARCHAR(40) NOT NULL,
cust_no INT,
FOREIGN KEY(seller_id) REFERENCES seller(seller_no),
FOREIGN KEY(cust_no) REFERENCES customer(cust_id)
);
"""
connection = create_db_connection("localhost", "root", pw, "agency")#Establishing db connection
execute_query(connection, create_property_table)##Sendign queries to be executed
execute_query(connection, create_customer_table)
execute_query(connection, create_agency_table)
execute_query(connection, create_enlists_table)
execute_query(connection, create_agent_table)
execute_query(connection, create_assigns_table)
execute_query(connection, create_seller_table)
execute_query(connection, create_searches_table)
execute_query(connection, create_sells_table)
```

```python
def read_query(connection, query):# This method will execute query and return the results
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")     
```
